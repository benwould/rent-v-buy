```python
import tkinter as tk
from tkinter import ttk
import app
  
 
LARGEFONT =("Verdana", 35)
  
class tkinterApp(tk.Tk):
     
    # __init__ function for class tkinterApp
    def __init__(self, *args, **kwargs):
         
        # __init__ function for class Tk
        tk.Tk.__init__(self, *args, **kwargs)
         
        # creating a container
        container = tk.Frame(self) 
        container.pack(side = "top", fill = "both", expand = True)
  
        container.grid_rowconfigure(0, weight = 1)
        container.grid_columnconfigure(0, weight = 1)
  
        # initializing frames to an empty array
        self.frames = {} 
  
        # iterating through a tuple consisting
        # of the different page layouts
        for F in (StartPage, Page1, Page2):
  
            frame = F(container, self)
  
            # initializing frame of that object from
            # startpage, page1, page2 respectively with
            # for loop
            self.frames[F] = frame
  
            frame.grid(row = 0, column = 0, sticky ="nsew")
  
        self.show_frame(StartPage)
  
    # to display the current frame passed as
    # parameter
    def show_frame(self, cont):
        frame = self.frames[cont]
        frame.tkraise()
  
# first window frame startpage
  
class StartPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        
        # label of frame Layout 2
        label = ttk.Label(self, text ="Rent vs Buy: User Data", font = LARGEFONT)
         
            
        my_rent_or_own = tk.StringVar(self, name="Rent or Own",value='Rent')
        my_city = tk.StringVar(self, name="My City",value='')
        my_monthly_payment = tk.IntVar(self, name="My Monthly Payment", value=100000)
        my_income = tk.IntVar(self, name="My Monthly Income($)", value=10000)
        my_home_value = tk.StringVar(self, name="My Home Value($)", value=660000)
        my_equity = tk.StringVar(self, name="My Equity($)", value=330000)
        my_interest_rate = tk.DoubleVar(self, name="My Interest Rate %", value=0.035)
        my_tax_rate = tk.DoubleVar(self, name="My Tax Rate %", value=0.01)
        
        label = tk.Label( self , text = "What city do you live in?" )
        label.grid(row=0,column=0)
        cities = app.get_city_list()
# Create Dropdown menu
        drop_cities = tk.OptionMenu( self , my_city , *cities )
        drop_cities.grid(row=0,column=1)
###### Rent or own
        label_ro = tk.Label( self , text = "Do you rent or own?" )
        label_ro.grid(row=1,column=0)
        ro = ["Rent","Own"]
        drop_ro = tk.OptionMenu( self , my_rent_or_own , *ro )
        drop_ro.grid(row=1,column=1)

###### Monthly Payment
        label_mp = tk.Label( self , text = "What is your monthly payment?" )
        label_mp.grid(row=2,column=0)
        entry_mp = tk.Entry(self,width=10,borderwidth=5)
        entry_mp.grid(row=2,column=1)

###### Monthly Income
        label_inc = tk.Label( self , text = "What is your monthly income?" )
        label_inc.grid(row=3,column=0)
        entry_inc = tk.Entry(self,width=10,borderwidth=5)
        entry_inc.grid(row=3,column=1)

###### Home Value
        label_hv = tk.Label( self , text = "What is your home value?" )
        label_hv.grid(row=4,column=0)
        entry_hv = tk.Entry(self,width=10,borderwidth=5)
        entry_hv.grid(row=4,column=1)

###### Equity
        label_eq = tk.Label( self , text = "What is your equity?" )
        label_eq.grid(row=5,column=0)
        entry_eq = tk.Entry(self,width=10,borderwidth=5)
        entry_eq.grid(row=5,column=1)

###### Interest Rate
        label_ir = tk.Label( self , text = "What is your interest rate?" )
        label_ir.grid(row=6,column=0)
        entry_ir = tk.Entry(self,width=10,borderwidth=5)
        entry_ir.grid(row=6,column=1)

        ###### Tax Rate
        label_tr = tk.Label( self , text = "What is your annual tax rate?" )
        label_tr.grid(row=7,column=0)
        entry_tr = tk.Entry(self,width=10,borderwidth=5)
        entry_tr.grid(row=7,column=1)

        def print_data():
            city=my_city.get()
            ownership=my_rent_or_own.get()
            my_monthly_payment = entry_mp.get()
            my_income = entry_inc.get()
            gmy_income = entry_inc.get()
            my_home_value = entry_hv.get()
            my_equity = entry_eq.get()
            my_interest_rate = entry_ir.get()
            my_tax_rate = entry_tr.get()
            label_data = tk.Label(self,text=f"""My Data:
                                     city:  {city}
                                     ownership:  {ownership}
                                     monthly payment: {my_monthly_payment}
                                     monthly income: {my_income}
                                     home value: {my_home_value}
                                     equity: {my_equity}
                                     interest rate: {my_interest_rate}
                                     tax rate: {my_tax_rate}
                                    """)
            label_data.grid(row=10,column=0,padx=0)

        def finished():
            self.destroy()
  

#label.grid(row = 0, column = 4, padx = 10, pady = 10)
  
        button1 = ttk.Button(self, text ="Page 1",
        command = lambda : controller.show_frame(Page1))
         
 # putting the button in its place by
 # using grid
        button1.grid(row = 9, column = 0, padx = 10, pady = 10)
  
        ## button to show frame 2 with text layout2
        button2 = ttk.Button(self, text ="Page 2",
        command = lambda : controller.show_frame(Page2))
     
        # putting the button in its place by
        # using grid
        button2.grid(row = 9, column = 1, padx = 10, pady = 10)
    # Create button, it will change label text
        button = ttk.Button(self
                         , text = "View My Data"
                         , command = print_data)
        button.grid(row=11,column=0,pady=0)
        button = ttk.Button( self
                             , text = "I'm done with data entry"
                             , command = finished)
        button.grid(row=11,column=1,pady=0)

        gmy_monthly_payment = entry_mp.get()
        my_income = entry_inc.get()
        my_home_value = entry_hv.get()
        my_equity = entry_eq.get()
        my_interest_rate = entry_ir.get()
        my_tax_rate = entry_tr.get()
    # Execute tkinter
        
    
  
          
  
  
# second window frame page1
class Page1(tk.Frame):
     
    def __init__(self, parent, controller):
        
         
        tk.Frame.__init__(self, parent)
        self.label = tk.Label(self, text ="Page 1", font = LARGEFONT)
        self.label.grid(row = 0, column = 4, padx = 10, pady = 10)
  
        # button to show frame 2 with text
        # layout2
        button1 = ttk.Button(self, text ="StartPage",
                            command = lambda : controller.show_frame(StartPage))
     
        # putting the button in its place
        # by using grid
        button1.grid(row = 1, column = 1, padx = 10, pady = 10)
  
        # button to show frame 2 with text
        # layout2
        button2 = ttk.Button(self, text ="Page 2",
                            command = lambda : controller.show_frame(Page2))
     
        # putting the button in its place by
        # using grid
        button2.grid(row = 2, column = 1, padx = 10, pady = 10)
  
  
  
# third window frame page2
class Page2(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.label = ttk.Label(self, text ="Page 2", font = LARGEFONT)
        self.label.grid(row = 0, column = 4, padx = 10, pady = 10)
  
        # button to show frame 2 with text
        # layout2
        button1 = ttk.Button(self, text ="Page 1",
                            command = lambda : controller.show_frame(Page1))
     
        # putting the button in its place by
        # using grid
        button1.grid(row = 1, column = 1, padx = 10, pady = 10)
  
        # button to show frame 3 with text
        # layout3
        button2 = ttk.Button(self, text ="Startpage",
                            command = lambda : controller.show_frame(StartPage))
     
        # putting the button in its place by
        # using grid
        button2.grid(row = 2, column = 1, padx = 10, pady = 10)
  
  
# Driver Code
app = tkinterApp()
app.mainloop()
```
